#Tokenization

####Sentence Segmentation


```python
sentences = nltk.sent_tokenize(paragraph) #used to split a paragraph of text into a list of individual sentences
```

```python
print(len(sentences))
```

```python
sentences
```

####Word Tokenization

```python
words = nltk.word_tokenize(paragraph) #used to break down a paragraph of text into a list of individual words
```

```python
print(len(words))
```

```python
words
```

```python
from nltk.tokenize import TreebankWordTokenizer
treebankwords = TreebankWordTokenizer()
treebankwords.tokenize(paragraph)
```

```python
print(len(treebankwords.tokenize(paragraph)))
```

```python
#WordPunctTokenizer splits text into words and treats punctuation marks as individual tokens.
from nltk.tokenize import WordPunctTokenizer
wordpunctwords = WordPunctTokenizer()
wordpunctwords.tokenize(paragraph)
```

```python
print(len(wordpunctwords.tokenize(paragraph)))
```

```python
#splits the paragraph into words, excluding punctuation marks
from nltk.tokenize import RegexpTokenizer
regexpwords = RegexpTokenizer(r'\w+')
regexpwords.tokenize(paragraph)
```

```python
print(len(regexpwords.tokenize(paragraph)))
```

```python
#analyze the frequency of words in a text corpus
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist.most_common(5)
```

```python
import matplotlib.pyplot as plt

fdist.plot(10)
```

####To remove punctuation marks

```python
words_no_punc = [] #empty list to store words

for w in words: #to remove punctuation marks
  if w.isalpha():
    words_no_punc.append(w.lower())

words_no_punc
```

###Stemming

```python
from nltk.stem import PorterStemmer
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

stopwords.fileids()
```

```python
stemmer = PorterStemmer()
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=''.join(words)
```

```python
words
```

#Lemmatization

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

sentences = nltk.sent_tokenize(paragraph)
lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i]=''.join(words)
```

```python
words
```

```python
from nltk.stem import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["study","studying","studies","studied"]

for w in word_list:
  print(lemma.lemmatize(w,pos='v'))
```
