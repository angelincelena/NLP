```python
paragraph = """Music is a universal language that transcends cultures and generations.
It's a beautiful tapestry woven from sound, combining elements like melody (pleasing sequences of notes),
rhythm (organized patterns of beats and accents), and harmony (the way different pitches blend together).
Music can evoke a wide range of emotions, from joy and excitement to sadness and contemplation.
It can energize us, soothe us, and even transport us to different times and places."""
```

#Sentence Tokenization

```python
sentences = nltk.sent_tokenize(paragraph)
sentences
```

#Word Tokenization

```python
words = nltk.word_tokenize(paragraph)
words
```

```python
from nltk.tokenize import TreebankWordTokenizer
treebankwords = TreebankWordTokenizer()
treebankwords.tokenize(paragraph)
```

```python
from nltk.tokenize import WordPunctTokenizer
wordpunctwords = WordPunctTokenizer()
wordpunctwords.tokenize(paragraph)
```

```python
from nltk.tokenize import RegexpTokenizer
regexpwords = RegexpTokenizer(r'\w+')
regexpwords.tokenize(paragraph)
```

#Analyze the frequency of words in a text corpus

```python
from nltk.probability import FreqDist
fdist = FreqDist(words)
fdist.most_common(5)
```

```python
import matplotlib.pyplot as plt

fdist.plot(10)
```

#To remove punctuation marks

```python
words_no_punc = [] #empty list to store words

for w in words: #to remove punctuation marks
  if w.isalpha():
    words_no_punc.append(w.lower())

words_no_punc
```

#Removing Stopwords

```python
nltk.download('stopwords')
```

```python
#filters out the commonly occurring words considered to have low meaning on their own
from nltk.corpus import stopwords

#list stopwords
stopwords = set(stopwords.words('english'))
print(stopwords)
```

```python
def remove_stopwords(paragraph):
  words = [word.lower() for word in nltk.word_tokenize(paragraph) if word not in stopwords]
  return ' '.join(words)

cleaned_paragraph = remove_stopwords(paragraph)
print(cleaned_paragraph)
```

#Word Cloud

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

colormap = 'plasma'

wordcloud = WordCloud(width=800, height=400, background_color='grey', colormap=colormap).generate(cleaned_paragraph)

plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

# Load the mask image
mask = np.array(Image.open('/content/image 1.jpg'))

colormap = 'plasma'

# Generate the word cloud with the mask
wordcloud = WordCloud(width=800, height=400, background_color='grey', mask=mask, colormap=colormap).generate(cleaned_paragraph)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
